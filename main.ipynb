# **Preprocessing**

## 1. Tokenisasi

```python
def hapusTandaBaca(text):
    return re.sub(r'[^\w\s]', '', text)

def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

def remove_spasi(text):
    return " ".join(text.split())

def remove(text):
    remove_angka = re.sub('[0-9]+', '', text)
    return remove_angka
```

```python
df["d1(0)"] = df["d1(0)"] .apply(lambda x: hapusTandaBaca(x)) \
                          .apply(lambda x: x.lower()) \
                          .apply(lambda x: remove_emoji(x)) \
                          .apply(lambda x: remove_spasi(x)) \
                          .apply(lambda x: remove(x)) \
                          .apply(lambda x: x.strip())
                          


df["d2(1)"] = df["d2(1)"] .apply(lambda x: hapusTandaBaca(x)) \
                          .apply(lambda x: x.lower()) \
                          .apply(lambda x: remove_emoji(x)) \
                          .apply(lambda x: remove_spasi(x)) \
                          .apply(lambda x: remove(x)) \
                          .apply(lambda x: x.strip())
```

# **Term Weighting (Pembobotan Kata)**

```python
def hitung_kemunculan_kata(list_kata):
  jumlah_kata = {}
  for kata in list_kata:
    if kata not in jumlah_kata:
      jumlah_kata[kata] = 0
    jumlah_kata[kata] += 1

  return jumlah_kata
```
